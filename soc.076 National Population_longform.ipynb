```python
import pandas as pd
import requests

url = 'https://esa.un.org/unpd/wup/Download/Files/WUP2018-F18-Total_Population_Annual.xls'
df1 = pd.read_excel(url, sheet_name = 0)
df1.head()
```

```python
#change the header
pop = df1.iloc[14:]
pop2.columns = pop.iloc[0]

#drop the header row
pop3 = pop2.drop(pop2.index[[0]])
pop3.rename(columns={pop3.columns[3]: 'code' ,'Region, subregion, country or area': 'country'}, inplace=True)
pop3
```

```python
#get columns
col = pop2.columns.get_values()
col2 = col.tolist()
del col2[:5]
col2
```

```python
pop3.head()
```

```python
#convert to longform
pop4 = pd.melt(pop3, id_vars=['code', 'country'], value_vars=list(col2))
pop5 = pop4.rename(columns={14: 'year'})
pop5.head()
```

```python
#export to CSV
pop5.to_csv('soc.076_Nat_pop_long')
```
